```python
df_country = pd.read_excel('Country-Code.xlsx')
```

```python
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df, df_country, on = 'Country Code', how = 'left')
```

```python
final_df.head()
```

```python
final_df.dtypes
```

```python
country_name = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3], labels = country_name[:3], autopct='%1.2f%%')
```

Observation: Maximum business of Zomato is from India after that United States and then United Kingdom

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns = {0: 'Rating count'})
```

```python
ratings
```

##### Observation:
######        1) When rating is between 4.5 to 4.9 --> Excellent
######        2) When rating is between 4.0 to 4.4 --> Very Good
######        3) When rating is between 3.5 to 3.9 --> Good
######        4) When rating is between 2.5 to 3.4 --> Average
######        5) When rating is between 1.8 to 2.4 --> Poor
######        6) When rating is  0.0 --> Not rated

```python
ratings.head()
```

```python
clr = ['#E0E0E0', 'Red', 'Orange', 'Yellow', 'Green', '#006400']
```

```python
sns.barplot(x = "Aggregate rating", y = "Rating count", data = ratings, hue = 'Rating color', palette = clr)
```

#### Observation:

    1) Not rated count is very high
    2) Maximum number of ratings are between 2.5 to 3.4

```python
sns.countplot(x = 'Rating color', data = ratings, hue = 'Rating color', palette = clr)
```

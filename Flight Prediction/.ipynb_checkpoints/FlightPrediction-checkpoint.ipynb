# Flight Price Prediction (EDA + Feature Engineering)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df = pd.read_excel("Data_Train.xlsx")
train_df.head()
```

```python
test_df = pd.read_excel("Test_set.xlsx")
test_df.head()
```

```python
final_df = pd.concat([train_df, test_df])
final_df.head()
```

```python
final_df.info()
```

```python
final_df[['Date', 'Month', 'Year']] = final_df['Date_of_Journey'].str.split('/', expand=True)
final_df[['Date', 'Month', 'Year']] = final_df[['Date', 'Month', 'Year']].astype(int)
final_df.drop('Date_of_Journey', axis = 1, inplace = True)
```

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
final_df[['Arrival_hour', 'Arrival_minutes']] = final_df['Arrival_Time'].str.split(':', expand=True)
final_df[['Arrival_hour', 'Arrival_minutes']] = final_df[['Arrival_hour', 'Arrival_minutes']].astype(int)
final_df.drop('Arrival_Time', axis = 1, inplace = True)
```

```python
final_df[['Dep_hour', 'Dep_minutes']] = final_df['Dep_Time'].str.split(':', expand=True)
final_df[['Dep_hour', 'Dep_minutes']] = final_df[['Dep_hour', 'Dep_minutes']].astype(int)
final_df.drop('Dep_Time', axis = 1, inplace = True)
```

```python
final_df.head()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'nan':1})
final_df.drop('Route',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
final_df['Duration_Minutes'] = (
    final_df['Duration'].str.extract(r'(?:(\d+)h)?\s*(?:(\d+)m)?')
    .fillna(0)
    .astype(int)
    .agg(lambda x: x[0]*60 + x[1], axis=1)
)
final_df.drop('Duration', axis = 1, inplace = True)
```

```python
final_df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
final_df['Airline']=labelencoder.fit_transform(final_df['Airline'])
final_df['Source']=labelencoder.fit_transform(final_df['Source'])
final_df['Destination']=labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info']=labelencoder.fit_transform(final_df['Additional_Info'])
```
